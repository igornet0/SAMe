```python
import numpy as np

def df_column_stats(df):
    stats = []

    for col in df.columns:
        col_data = df[col]
        
        stat = {
            "column": col,
            "dtype": col_data.dtype,
            "missing_count": col_data.isna().sum(),
            "missing_pct": col_data.isna().mean() * 100,
            "unique_count": col_data.nunique()
        }
        
        # Для строк — длины
        if col_data.dtype == object or pd.api.types.is_string_dtype(col_data):
            lengths = col_data.dropna().astype(str).map(len)
            stat.update({
                "avg_length": lengths.mean(),
                "min_length": lengths.min(),
                "max_length": lengths.max()
            })
        
        # Для чисел — статистика
        elif pd.api.types.is_numeric_dtype(col_data):
            stat.update({
                "mean": col_data.mean(),
                "min": col_data.min(),
                "max": col_data.max(),
                "std": col_data.std()
            })
        
        stats.append(stat)
    
    return pd.DataFrame(stats)

# Пример использования
stats_df = df_column_stats(df_t)
```

```python
def token_split(token:str):
   if len(token) >= 3:
      if "," in token or "." in token:
         if not token.replace(".", "").replace(",", "").isdigit():
            token = token.replace(",", " ").replace(".", " ")

      if "(" in token or ")" in token:
         if not token.replace("(", "").replace(")", "").isdigit():
            token = token.replace("(", " ").replace(")", " ")

      if ("(" in token or ")" in token) and "-" in token:
         token = token.replace("-", " - ")

      if "/" in token and not(token.replace("/", "").isdigit()):
         token =  token.replace("/", " / ")
      
   return token.split()
```
