```python
# –î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏—è —Ä–∞–±–æ—Ç—ã same_clear
print("üß™ –¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ same_clear –º–æ–¥—É–ª–µ–π")
print("=" * 50)

# –¢–µ—Å—Ç–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ
test_texts = [
    '–ë–æ–ª—Ç –ú10—Ö50 <b>–ì–û–°–¢</b> 7798-70 &nbsp; —Å –ø–æ–∫—Ä—ã—Ç–∏–µ–º',
    '–ì–∞–π–∫–∞ –ú10 –ì–û–°–¢ 5915-70 –æ—Ü–∏–Ω–∫–æ–≤–∞–Ω–Ω–∞—è',
    '–®–∞–π–±–∞ 10 –ì–û–°–¢ 11371-78 –ø–ª–æ—Å–∫–∞—è',
    '–í–∏–Ω—Ç –ú8—Ö30 DIN 912 —Å –≤–Ω—É—Ç—Ä–µ–Ω–Ω–∏–º —à–µ—Å—Ç–∏–≥—Ä–∞–Ω–Ω–∏–∫–æ–º',
    '–¢—Ä—É–±–∞ —Å—Ç–∞–ª—å–Ω–∞—è √ò57—Ö3.5 –ì–û–°–¢ 8732-78'
]

# 1. –û—á–∏—Å—Ç–∫–∞ —Ç–µ–∫—Å—Ç–∞
print("\n1Ô∏è‚É£ –û—á–∏—Å—Ç–∫–∞ —Ç–µ–∫—Å—Ç–∞:")
cleaner = TextCleaner()
for i, text in enumerate(test_texts[:3], 1):
    result = cleaner.clean_text(text)
    print(f"{i}. –ò—Å—Ö–æ–¥–Ω—ã–π: {result['raw']}")
    print(f"   –û—á–∏—â–µ–Ω–Ω—ã–π: {result['normalized']}")
    print()

# 2. –ò–∑–≤–ª–µ—á–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
print("\n2Ô∏è‚É£ –ò–∑–≤–ª–µ—á–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤:")
extractor = RegexParameterExtractor()
for i, text in enumerate(test_texts[:3], 1):
    params = extractor.extract_parameters(text)
    print(f"{i}. –¢–µ–∫—Å—Ç: {text}")
    print(f"   –ü–∞—Ä–∞–º–µ—Ç—Ä—ã ({len(params)}):")
    for param in params[:3]:  # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º –ø–µ—Ä–≤—ã–µ 3
        print(f"     - {param.name}: {param.value} ({param.parameter_type.value})")
    print()
```

```python
# –î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏—è —Ä–∞–±–æ—Ç—ã same_search
print("üîç –¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ same_search –º–æ–¥—É–ª–µ–π")
print("=" * 50)

# –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –ø–æ–∏—Å–∫–∞
documents = [
    '–ë–æ–ª—Ç –ú10—Ö50 –ì–û–°–¢ 7798-70 –æ—Ü–∏–Ω–∫–æ–≤–∞–Ω–Ω—ã–π',
    '–ì–∞–π–∫–∞ –ú10 –ì–û–°–¢ 5915-70 —à–µ—Å—Ç–∏–≥—Ä–∞–Ω–Ω–∞—è',
    '–®–∞–π–±–∞ 10 –ì–û–°–¢ 11371-78 –ø–ª–æ—Å–∫–∞—è',
    '–í–∏–Ω—Ç –ú8—Ö30 DIN 912 —Å –≤–Ω—É—Ç—Ä–µ–Ω–Ω–∏–º —à–µ—Å—Ç–∏–≥—Ä–∞–Ω–Ω–∏–∫–æ–º',
    '–ë–æ–ª—Ç –ú12—Ö60 –ì–û–°–¢ 7798-70 —Å –ø–æ–ª–Ω–æ–π —Ä–µ–∑—å–±–æ–π',
    '–ì–∞–π–∫–∞ –ú12 –ì–û–°–¢ 5915-70 –æ—Ü–∏–Ω–∫–æ–≤–∞–Ω–Ω–∞—è',
    '–®–ø–∏–ª—å–∫–∞ –ú10—Ö100 –ì–û–°–¢ 22032-76',
    '–í–∏–Ω—Ç –ú6—Ö20 DIN 912 –Ω–µ—Ä–∂–∞–≤–µ—é—â–∏–π'
]
doc_ids = [str(i) for i in range(1, len(documents) + 1)]

print(f"üìä –ü–æ–¥–≥–æ—Ç–æ–≤–ª–µ–Ω–æ {len(documents)} –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤ –¥–ª—è –∏–Ω–¥–µ–∫—Å–∞—Ü–∏–∏")

# 1. –ù–µ—á–µ—Ç–∫–∏–π –ø–æ–∏—Å–∫
print("\n1Ô∏è‚É£ –ù–µ—á–µ—Ç–∫–∏–π –ø–æ–∏—Å–∫ (Fuzzy Search):")
fuzzy_engine = FuzzySearchEngine()
fuzzy_engine.fit(documents, doc_ids)

query = "–±–æ–ª—Ç –º10"
results = fuzzy_engine.search(query, top_k=3)
print(f"–ó–∞–ø—Ä–æ—Å: '{query}'")
print("–†–µ–∑—É–ª—å—Ç–∞—Ç—ã:")
for i, result in enumerate(results, 1):
    print(f"  {i}. {result['content']} (score: {result['score']:.3f})")

# 2. –°–µ–º–∞–Ω—Ç–∏—á–µ—Å–∫–∏–π –ø–æ–∏—Å–∫ (–µ—Å–ª–∏ –¥–æ—Å—Ç—É–ø–µ–Ω)
print("\n2Ô∏è‚É£ –°–µ–º–∞–Ω—Ç–∏—á–µ—Å–∫–∏–π –ø–æ–∏—Å–∫ (Semantic Search):")
try:
    semantic_engine = SemanticSearchEngine()
    semantic_engine.fit(documents, doc_ids)
    
    query = "–∫—Ä–µ–ø–µ–∂–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç"
    results = semantic_engine.search(query, top_k=3)
    print(f"–ó–∞–ø—Ä–æ—Å: '{query}'")
    print("–†–µ–∑—É–ª—å—Ç–∞—Ç—ã:")
    for i, result in enumerate(results, 1):
        print(f"  {i}. {result['content']} (score: {result['score']:.3f})")
except Exception as e:
    print(f"‚ö†Ô∏è –°–µ–º–∞–Ω—Ç–∏—á–µ—Å–∫–∏–π –ø–æ–∏—Å–∫ –Ω–µ–¥–æ—Å—Ç—É–ø–µ–Ω: {e}")
    print("üí° –î–ª—è —Ä–∞–±–æ—Ç—ã —Ç—Ä–µ–±—É–µ—Ç—Å—è —É—Å—Ç–∞–Ω–æ–≤–∫–∞ sentence-transformers")
```

```python
# –ü–æ–ª–Ω—ã–π –ø–∞–π–ø–ª–∞–π–Ω –æ–±—Ä–∞–±–æ—Ç–∫–∏
print("üîÑ –ü–æ–ª–Ω—ã–π –ø–∞–π–ø–ª–∞–π–Ω –æ–±—Ä–∞–±–æ—Ç–∫–∏")
print("=" * 50)

# –ò—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
raw_text = '–ë–æ–ª—Ç –ú10—Ö50 <b>–ì–û–°–¢</b> 7798-70 &nbsp; –æ—Ü–∏–Ω–∫–æ–≤–∞–Ω–Ω—ã–π'
print(f"üìù –ò—Å—Ö–æ–¥–Ω—ã–π —Ç–µ–∫—Å—Ç: {raw_text}")

# –®–∞–≥ 1: –û—á–∏—Å—Ç–∫–∞ —Ç–µ–∫—Å—Ç–∞ (same_clear)
cleaner = TextCleaner()
cleaned_result = cleaner.clean_text(raw_text)
cleaned_text = cleaned_result['normalized']
print(f"üßπ –ü–æ—Å–ª–µ –æ—á–∏—Å—Ç–∫–∏: {cleaned_text}")

# –®–∞–≥ 2: –ò–∑–≤–ª–µ—á–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ (same_clear)
extractor = RegexParameterExtractor()
parameters = extractor.extract_parameters(cleaned_text)
print(f"üîç –ò–∑–≤–ª–µ—á–µ–Ω–æ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤: {len(parameters)}")
for param in parameters[:3]:
    print(f"   - {param.name}: {param.value}")

# –®–∞–≥ 3: –ü–æ–∏—Å–∫ –∞–Ω–∞–ª–æ–≥–æ–≤ (same_search)
fuzzy_engine = FuzzySearchEngine()
catalog = [
    '–ë–æ–ª—Ç –ú10—Ö50 –ì–û–°–¢ 7798-70 –æ—Ü–∏–Ω–∫–æ–≤–∞–Ω–Ω—ã–π',
    '–ë–æ–ª—Ç –ú10—Ö60 –ì–û–°–¢ 7798-70 –æ—Ü–∏–Ω–∫–æ–≤–∞–Ω–Ω—ã–π', 
    '–ë–æ–ª—Ç –ú12—Ö50 –ì–û–°–¢ 7798-70 –æ—Ü–∏–Ω–∫–æ–≤–∞–Ω–Ω—ã–π',
    '–ì–∞–π–∫–∞ –ú10 –ì–û–°–¢ 5915-70 –æ—Ü–∏–Ω–∫–æ–≤–∞–Ω–Ω–∞—è'
]
fuzzy_engine.fit(catalog, [str(i) for i in range(len(catalog))])
search_results = fuzzy_engine.search(cleaned_text, top_k=3)

print(f"üéØ –ù–∞–π–¥–µ–Ω–æ –∞–Ω–∞–ª–æ–≥–æ–≤: {len(search_results)}")
for i, result in enumerate(search_results, 1):
    print(f"   {i}. {result['content']} (score: {result['score']:.3f})")

# –®–∞–≥ 4: –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –∫ —ç–∫—Å–ø–æ—Ä—Ç—É (same_api)
export_data = pd.DataFrame([
    {
        'Raw_Name': raw_text,
        'Cleaned_Name': cleaned_text,
        'Parameters_Count': len(parameters),
        'Best_Match': search_results[0]['content'] if search_results else '–ù–µ –Ω–∞–π–¥–µ–Ω–æ',
        'Best_Score': search_results[0]['score'] if search_results else 0.0
    }
])

print(f"\nüìä –†–µ–∑—É–ª—å—Ç–∞—Ç –æ–±—Ä–∞–±–æ—Ç–∫–∏:")
print(export_data.to_string(index=False))

print("\n‚úÖ –ü–∞–π–ø–ª–∞–π–Ω –≤—ã–ø–æ–ª–Ω–µ–Ω —É—Å–ø–µ—à–Ω–æ!")
print("üéâ –í—Å–µ –º–æ–¥—É–ª–∏ —Ä–∞–±–æ—Ç–∞—é—Ç –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ!")
```
